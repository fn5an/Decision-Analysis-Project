## Data Simulation

##### Importing required libraries

```python
import pandas as pd
import os, sys
import random
```

##### Read a profile CSV file which contains participants' ID numbers, genders, and age group .

```python
data = pd.read_csv('/Users/fatemenikseresht/Dropbox/Bias Data/Simulation.csv')
data = data.set_index('Participant-ID')
#data
```

##### Create a data frame which contains five different emotional status for each participant. Each participant will experience all of this emotional status but in a random sequence and with different probabilities. 


```python
status = ["sad", "tired", "tense", "exhausted", "angry"]
emotion = pd.DataFrame()
for x in range(60):
    emotion = pd.concat([emotion, pd.DataFrame(status).transpose()], axis=0)
    random.shuffle(status)
    
emotion = emotion.rename(columns={0:'Status1',1:'Status2',2:'Status3',3:'Status4',4:'Status5'})    
o = pd.Series(range(1,61))
emotion.index=o
#emotion
```

##### Create participants preferences dateframe that shows participant feedback preferences with scores from 1-5 

```python
score = ["1", "2", "3", "4", "5"]
prefrences = pd.DataFrame()
for x in range(60):
    prefrences = pd.concat([prefrences, pd.DataFrame(score).transpose()], axis=0)
    random.shuffle(score)
    
prefrences = prefrences.rename(columns={0:'Listen to Music',1:'Drink Coffee',2:'Drink Water',3:'Sleep/Have a rest',4:'Exercise'})    
o = pd.Series(range(1,61))
prefrences.index=o
#prefrences
```

##### Concatenate profiles dataframe, preferences dataframe, and emotional statuses dataframe.

```python
df= pd.concat([data, prefrences, emotion], axis=1 )
df
```
